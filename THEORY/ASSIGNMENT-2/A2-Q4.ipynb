```python
import pandas as pd
import numpy as np
```

```python
tdf = pd.read_csv("titanic_cleaned.csv")
fdf = pd.read_csv("ticket_fares.csv")
```

```python
pd.merge(tdf,fdf,on='Ticket')
```

```python
ageBins =[0,12,19,59,100]
ageLabels= ['Child','Young Adult','Adult','Senior']
tdf["AgeGroup"] = pd.cut(tdf["Age"],bins= ageBins,labels=ageLabels,right=True)
```

```python
tdf.groupby(['Sex','AgeGroup'])['Survived'].mean()
```

```python
with open("report.txt","w") as f:
    f.write("Hypothesis A\n")
    f.write("The numbers clearly show that women and children were given priority during the Titanic rescue. Most women survived, with survival rates between 74% to 100%, while only about 10% to 17%  men survived. Children also had better chances of survival, with around 50% to 57% were saved. ")
```

```python
tdf.groupby("Pclass")["Survived"].mean()
```

```python
fareLabels = ['Low','Medium','High','VeryHigh']
tdf["FareBin"] = pd.qcut(tdf["Fare"],q=4,labels=fareLabels)
```

```python
tdf.groupby("FareBin")["Survived"].mean()
```

```python
with open("report.txt","a") as f:
    f.write("\nHypothesis B\n")
    f.write("The data clearly shows that wealthier passengers had a higher chance of survival on the Titanic. Passengers in the lowest fare group had only about 19% survival, while those in the highest fare group had around 58%. This suggests that people who paid more, likely traveling in higher classes, were given better access to lifeboats and safety, showing a clear link between wealth and survival.")
```
