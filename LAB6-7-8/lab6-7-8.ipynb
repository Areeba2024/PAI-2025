# Question 4

```python
def analysis(rating):
    if rating >= 4:
        return "Positive"
    elif rating == 2.5:
        return "Neutral"
    else: 
        return "Negative"
df['Sentiments'] = df['rating'].apply(analysis)
df
```

```python
count = df['Sentiments'].value_counts()

plt.pie(count, labels=count.index, autopct='%1.1f%%')
plt.show()
```

# Question 5

```python
def findModel(price,rating):
    if price > df["price_usd"].mean() and rating < df["rating"].mean():
        return 1
    else:
        return 0

df["Overpriced"] = df.apply(lambda row: findModel(row["price_usd"], row["rating"]), axis=1)
```

```python
plt.scatter(df['model'],df['Overpriced'])
plt.xlabel("Models")
plt.ylabel("Overpriced category")
plt.xticks(rotation = 45)
plt.show()
```
